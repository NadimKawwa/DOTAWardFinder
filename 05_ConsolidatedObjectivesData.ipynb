## Combine Using Left Join

Left = ward dataframe observer/sentry
right = objective dataframe

one to many situation

```python
#merge on match_id common key
df_obs_obj = df_obs.merge(right=df_obj,
                          how='left', 
                          left_on = 'match_id',
                          right_on = 'match_id'
                         )
#save
df_obs_obj.to_csv('df_obs_obj.csv',
                  index=False
                 )
```

```python
#take a look at missing values
df_obs_obj.isna().mean()
```

```python
#repeat for sentry 
df_sen_obj = df_sen.merge(right=df_obj,
                          how='left', 
                          left_on = 'match_id',
                          right_on = 'match_id'
                         )

df_sen_obj.to_csv('df_sen_obj.csv', 
                  index=False
                 )
```
